```python
#WAP to create an array of integer[2,4,6,8,10] and traverse the element
import array as arr
a=arr.array('i',[2,4,6,8,10])
print(a)
for i in range(0,len(a)):
  print(a[i],end=" ")
```

```python
# WAP to create a array of float element.insert 9.5 at index 2 and append 4.5,6.5 at the end of array
import array as arr
a=arr.array('d',{2,3,4,5})
for i in range(0,len(a)):
  print(a[i],end=" ")
a.insert(2,9.5)
a.extend([4.5,6.5])
print()
for i in range(0,len(a)):
  print(a[i],end=" ")
```

```python
#WAP to remove first occurence of 7 from integer array
import array as arr
a=arr.array('i',[1,4,7,3,4,5,7,8])
for i in range(0,len(a)):
  if a[i]==7:
    a.remove(7)
    break
for i in range(0,len(a)):
  print(a[i],end=" ")
```

```python
#write code to reverse an array at place
import array as arr
a=arr.array('i',[1,2,3,4,5])
start=0
end=len(a)-1
while start<=end:
  a[start],a[end]=a[end],a[start]
  start+=1
  end-=1

for i in a:
  print(i,end=" ")
```

```python
#find largest and smallest element in the array
#using build in function
import array as arr
a=arr.array('i',[1,2,3,4,5,6,7])
print("smallest element in array:",min(a))
print("largest element in array:",max(a))
```

```python
#find smallest and largest element in array
import array as arr
a=arr.array('i',[2,3,4,1,6,8,7,55])
small=a[0]
large=a[0]
for i in a:
  if i<small:
    small=i
  if i>large:
    large=i
print("smallest element of array:",small)
print("largest element of array:",large)
```
