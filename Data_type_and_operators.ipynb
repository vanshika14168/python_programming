```
#WAP to accept two  number from user and perform all arithmetic operations
a=float(input("enter a number:"))
b=float(input("enter number:"))
print(a,"+",b,"=",a+b)
print(a,"-",b,"=",a-b)
print(a,"//",b,"=",a//b)
print(a,"%",b,"=",a%b)
print(a,"*",b,"=",a*b)
```

```
#calculate area of circle,triangle,rectangle
r=float(input("enter radius:"))
print("area of circle is:",3.14*r*r)
l=int(input("enter length:"))
b=int(input("enter breath:"))
print("area of rectangle is:",l*b)
h=int(input("enter height:"))
b=int(input("enter base:"))
print("area of triangle is:",(b*h)//2)
```

```
#create restaurant order system in which input price of three food item and calculate total bill with 5% GST
a=int(input("enter the price of first food item:"))
b=int(input("enter the price of second food item:"))
c=int(input("enter the price of third food item:"))
t=a+b+c
print("total food bill:",t)
g=(t*5)//100
print("bill GST is:",g)
print("total bill with 5% GST:",t+g)
```

```
#convert temperature from celsius to fahrenheit
#F=(C*9//5)+32
c=int(input("enter the temperature in celsius:"))
F=(c*9//5)+32
print("convert temperature from celsius to fahrenheit:",F)
```

```
#compute simple interest
p=int(input("enter principal:"))
r=int(input("enter rate:"))
t=int(input("enter time:"))
S_I=(p*r*t)//100
print("simple interest is:",S_I)
```

```
#swapping of two number without using third variable
a=int(input("enter number:"))
b=int(input("enter number:"))
print("before swapping:",a,b)
a,b=b,a
print("after swapping",a,b)
```

```
#calulate total,average and percentage of 5 subject numbers
m1=int(input("enter the marks of first subject:"))
m2=int(input("enter the marks of second subject:"))
m3=int(input("enter the marks of third subject:"))
m4=int(input("enter the marks of fourth subject:"))
m5=int(input("enter the marks of fifth subject:"))
t=m1+m2+m3+m4+m5
print("total marks of five subject is:",t)
a=t//5
print("average of 5 subject:",a)
print("total percentage is:",a)
```

```
#take complex number input from user and print its real and imaginary part
c=complex(input("enter complex number:"))
print("real part of complex number is:",c.real)
print("imaginary part of complex number is:",c.imag)
```

```
#convert days into week and months
day=int(input("enter the number of days:"))
print("convert days from days to week:",day//7)
print("convert days from day to month:",day/30)
```

```
#WAP to split bill among three friends print amount paid by each friend and the remaining amount
p1=int(input("enter the price of first item:"))
p2=int(input("enter the price of second item:"))
p3=int(input("enter the price of third item:"))
f=int(input("enter the number of friends among split the bill:"))
t=p1+p2+p3
print("total bill:",t)
E=(p1+p2+p3)//f
print("bill pay by each friend:",E)
print("remaining bill:",(t-(E*f)))
```
