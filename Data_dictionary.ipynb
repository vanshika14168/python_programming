```python
#create and display dictionary
student={"name":"vanshika","age":20,"section":"data-science"}
print(student)
```

```python
#access specific value using keys
student={"name":"vanshika","age":20,"section":"data-science"}
print(student)
print(student.keys())
print(student.values())
print(student.items())
```

```python
#add and update item in dictionary
student={"name":"amit","age":20,"marks":90}
print(student)
student["section"]="data-science"
print(student)
student["age"]=23
print(student)
```

```python
#remove key value pair from dictionary
student={"name":"amit","age":29,"marks":99}
print(student)
student.pop("age")
print(student)
```

```python
#check if key exist in dictionary
student={"name":"sonu","age":22,"marks":98}
print(student)
if "age" in student:
  print("exist in student")
else:
  print("not exist in student")
```

```python
#use dictionary methods,key(),item(),value()
student={"name":"ashi","section":"A","grade":"A1"}
print(student)
print(student.keys())
print(student.values())
print(student.items())
```

```python
#demonstrate copy and clear method
student={"name":"ashi","section":"A1","marks":99}
print(student)
student1=student.copy()
print(student1)
student.clear()
print(student)
```

```python
#you are given name of student and their marks in test
"""write a program to :
store data in a dictionary
display student names with marks
find the student with the highest marks
"""

student={"ashi":99,"amit":89,"sonu":78,"mohit":56}
print(student)
print(student.items())
print("student with highest marks:",max(student.keys()))
```

```python
#create a dictionary of employee with name and their name and salary
""" display the average salary
    find the emloyee with salary above the average"""

employee={"ramu":9000,"sonu":7000,"rohit":8900,"shubham":9900}
print(employee)
avg_salary=sum(employee.values())/len(employee)
print("average salary:",avg_salary)
for name,salary in employee.items():
  if(salary>avg_salary):
    print("employee with salary above the average salary",(name,salary))
```

```python
"""store student marks of multiple subjects for a student in a nested dictionary
and compute average marks."""

student={"amit":{
    "maths":90,"science":85,"english":88,"computer":89
}}
print(student)
marks=student["amit"]
print(marks)
print("average marks:",sum(marks.values())/len(marks))
```

```python
#given a dictionary of item price apply 10% discount to items priced above 1000rs
item={"bag":300,"shoes":1500,"watch":2500,"bottle":300}
print(item)
for key,price in item.items():
  if(price>1000):
    item[key]=price-(price*0.1)
print(item)
```
