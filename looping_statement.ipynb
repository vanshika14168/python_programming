```python
#WAP to print sum of digit of a number
num=int(input("enter a number:"))
num1=num
sum=0
while num!=0:
  last_digit=num%10
  sum=sum+last_digit
  num=num//10
print(f"sum of {num1} is: {sum}")
```

```python
#WAP to print reverse of a number
num=int(input("enter a number:"))
num1=num
ans=0
while num!=0:
  last_digit=num%10
  ans=ans*10+last_digit
  num=num//10
print(f"reverse of a {num1} is :{ans}")
```

```python
#WAP to print count digit in a number
num=int(input("enter a number:"))
num1=num
count=0
while num!=0:
  num=num//10
  count=count+1
print(f"total number of digit in {num1} is:{count}")
```
