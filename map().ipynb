```python
#define map().give syntax with example
"""map() is a built-in python function used to apply a given function to each item
of an iterable(like list,tuple,etc)and return a map object containing result
syntax: map(function,iterable)"""

def square(x):
  return x*x

numbers=[1,2,3,4]
result=map(square,numbers)
print(list(result))
```
