```python
#create a tuple of colors.print its length and access the second and last element
color=("red","blue","green","pink","yellow","gold")
print(color)
print("length of tuple is:",len(color))
print("second element of tuple:",color[1])
print("last element of tuple:",color[-1])
```

```python
#create a tuple (1,2,3) and(4,5) perform concatenation&repeat(7,8) three times
t1=(1,2,3)
t2=(4,5)
print(t1+t2)
print((7,8)*3)
```

```python
#create a tuple of three element,where the last element is another tuple(7,8,9)
#and access the value 8 from it
t1=(1,2,(7,8,9))
print(t1[2][1])
```

```python
"""create a tuple(3,5,3,2,5,5,1) find :
the smallest value
the largest value
how many times 5 appears"""
tuple=(3,5,3,2,5,5,1)
print(tuple)
print("smallest value:",min(tuple))
print("largest value:",max(tuple))
count=0
for i in tuple:
  if i==5:
    count=count+1
print("5 appear :",count,"times")
```
