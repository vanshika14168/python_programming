```python
#two employe skill sets are:
"""E1={"python","sql",c++"}
   E2={"python","java","html"}
   write a program to find
   skill unique to E1
   skill unique to E2
   all skill required in the company"""

E1={"python","sql","c++"}
E2={"python","java","html"}
print("skill unique to E1:",E1-E2)
print("skill unique to E2:",E2-E1)
print("all common skills:",(E1 and E2))
print("all skills:",(E1 or E2))
```

```python
E1={"python","sql","c++"}
E2={"python","java","html"}

print("skill unique to E1:",E1.difference(E2))
print("skill unique to E2:",E2.difference(E1))
print("common skill:",E1.intersection(E2))
print("all skill:",E1.union(E2))
```

```python
#Give three sets representing student who play cricket,football,badminton.
#find student who play exactly two sports
cricket={"a","b","c","d","e"}
football={"b","c","f","g","e"}
badminton={"a","c","g","f"}
two_play=((cricket and football) or (football and badminton) or (badminton and cricket))-(cricket and football and badminton)
print("student who play exactly two play:",two_play)
```

```python
#you have two sets:
"""passed={"A","B","C","D"}
   Failed={"E","F","B"}
   find:students who pass only
   student who both passed and fail"""

passed={"A","B","C","D"}
failed={"E","F","B"}

passed_only=passed-failed
print("student who pass only:",passed_only)

both_passed_failed=passed.intersection(failed)
print("student who both passed and failed:",both_passed_failed)
```
