```python
#WAP to calculate electricity bill based on units consumed
#0-100 unit-->5rs per unit
#101-200 unit-->7rs per unit
#above 200 unit-->10rs per unit
unit=int(input("enter the units consumed:"))
if unit>=0 and unit<=100:
  print("total bill:",unit*5)
elif unit>100 and unit<=200:
  print("total bill",(100*5)+(unit-100)*7)
else:
  print("total bill:",(100*5)+(100*7)+(unit-200)*10)
```

```python
#check a triangle equilateral,isosceles,or scalene using its side length
s1=int(input("enter first side:"))
s2=int(input("enter second side:"))
s3=int(input("enter third side:"))
if s1==s2 and s1==s3 :
  print("Equilateral triangle")
elif (s1==s2) or (s2==s3) or (s3==s1):
  print("isoscele triangle")
else:
  print("scalene triangle")
```

```python
#library fine calculator ,input number of days
#0-5days-->5rs/day
#6-10days-->10rs/day
#10days-->20rs/day
#30days-->membership cancelled
day=int(input("enter number of days:"))
if day>=0 and day<=5:
  print("total fine:",day*5)
elif day>5 and day<=10:
  print("total fine:",day*10)
elif day>10 and day<=30:
  print("total fine:",day*20)
else:
  print("memebership cancelled")
```

```python
#WAP to implement a simple calculator using if-elif-else(add,subtract,multiply,divide)
num1=int(input("enter first number:"))
num2=int(input("enter second number:"))
operator=input("enter operator:")
if operator=='+':
  print(num1,"+",num2,"=",num1+num2)
elif operator=='-':
  print(num1,"-",num2,"=",num1-num2)
elif operator=='*':
  print(num1,"*",num2,"=",num1*num2)
elif operator=='//':
  print(num1,"//",num2,"=",num1//num2)
else:
  print("invalid operator")
```
