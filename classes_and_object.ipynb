```python
#create a class student with details and display them
class student:
  def __init__(self,name,age,roll_no):
    self.name=name
    self.age=age
    self.roll_no=roll_no
  def display(self):
    print("name is:",self.name)
    print("age is:",self.age)
    print("roll no. is:",self.roll_no)

S1=student("vanshike",20,67)
S1.display()
```

```python
#create a class employee and calculate salary
class employee:
  def __init__(self,name,salary):
    self.name=name
    self.salary=salary

  def calculate_salary(self):
    HRA=0.20*self.salary #house rent allowance
    DA=0.10*self.salary #dearness allowance
    total_salary=self.salary+HRA+DA
    return total_salary

  def display(self):
    print("employee name:",self.name)
    print("basic salary:",self.salary)
    print("total salary:",self.calculate_salary())

E1=employee("vamshika",100000)
E1.display()
```

```python
#create a class rectangle and find area and perimeter
class Rectangle:
  def __init__(self,length,width):
    self.length=length
    self.width=width

  def area(self):
    return self.length*self.width

  def perimeter(self):
    return 2*(self.length*self.width)

  def display(self):
    print("length is:",self.length)
    print("width is:",self.width)
    print("area is:",self.area())
    print("perimeter is:",self.perimeter())

R1=Rectangle(23,2)
R1.display()
```

```python
#create a class bank account,create function deposite and withdraw and show balance
class BankAccount:
  def __init__(self,name,balance):
    self.name=name
    self.balance=balance

  def deposite(self,amount):
    self.balance+=amount
    print("deopsite:",amount)

  def withdraw(self,amount):
    if amount<=self.balance:
      self.balance-=amount
      print("withdraw amount:",amount)
    else:
      print("insufficient amount")

  def show(self):
    print("account holder name:",self.name)
    print("current balance:",self.balance)

B=BankAccount("vanshika",565)
B.show()

B.deposite(500)
B.show()

B.withdraw(600)
B.show()
```

```python
#create a class book,accept and display book details
class Book:
  def __init__(self):
    self.title=""
    self.author=""
    self.price=0

  def accept(self):
    self.title=input("enter book title:")
    self.author=input("enter author name:")
    self.price=float(input("enter price:"))

  def display(self):
   print("\nBOOK DETAIS.....")
   print("Book title:",self.title)
   print("Book author:",self.author)
   print("Book price:",self.price)

B=Book()
B.accept()
B.display()
```

```python
#create a class movie and check if box collection is greater then 100 then it is hit else flop
class Movie:
  def __init__(self,movie_name,box_collection):
    self.movie_name=movie_name
    self.box_collection=box_collection

  def check_status(self):
    if self.box_collection>100:
      print("movie is hit")
    else:
      print("movie is flop")

M=Movie("dhurandar",800)
M.check_status()
```

```python
#create a shopping cart.add item and show total cost
class Shopping:
  def __init__(self):
    self.items=[]
    self.prices=[]

  def add_item(self,item,price):
    self.items.append(item)
    self.prices.append(price)
    print(item," add to cart")
  print()

  def display_price_item(self):
   for i in range(len(self.items)):
    print(self.items[i]," - ",self.prices[i])

  def total_price(self):
    price=sum(self.prices)
    print("\ntotal price is:",price)

S=Shopping()
S.add_item("pen",20)
S.add_item("choclate",70)
S.add_item("book",200)

print("\ndisplay price...")
S.display_price_item()
S.total_price()
```
